## 컬럼 추가 & 수정하기
컬럼 추가하는 데 쓰는 코드는 모두 동일하게 수정에도 적용

```python
train["Category"] = "Titanic"
train.head()
# 카테고리라는 컬럼을 추가하는데, 그 값이 Titanic으로 모두 같은 경우
```

```python
# 개수만 맞으면 들어간다.
train["Id"] = range(891)
train.head()

# 만약 저게 890,892 이어도 안들어감
```

```python
# 가족 숫자 구하기
train["Familysize"] = train["Parch"] + train["SibSp"] +1
train.head()
train[["Parch", "SibSp", "Familysize"]].head()
```

```python
#조건문 걸고 생성
train["France"] = train["Embarked"] == "C"
train["UK"] = train["Embarked"] != "C"

train[["Embarked", "France", "UK"]].head()

# 이 방식은 조건마다 컬럼이 추가됨 (단점)
```

```python
# 색인하는 방법대로
# train.loc[row_이 값은 조건문, column_검색해서 나타내고자 하는 컬럼]
# "Nationality" 라느 컬럼을 추가하는데, 여기에 위의 두 정보가 모두 포함됐으면 좋겠다 하는 경우
train.loc[train["Embarked"] == "C", "Nationality"] = "France"
train.loc[train["Embarked"] != "C", "Nationality"] = "UK"

train[["Embarked", "Nationality"]].head()
```

```python
train["high_fare"] = train["Fare"] > 30
train[["Fare", "high_fare"]].head()
```

```python
train.loc[train["Fare"] > 30, "high_fare"] = "high"
train[["Fare", "high_fare"]].head()
```

```python
mean_age = train["Age"].mean()
mean_age
```

```python
#bad -train.loc[train["Age"].isnull()]["Age"] = mean_age 
train.loc[train["Age"].isnull(), "Age"] = mean_age
train.head(10)
```
